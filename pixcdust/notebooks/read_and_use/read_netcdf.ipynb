# Simply read a single PixC netcdf


```python
from pixcdust.readers.netcdf import NcSimpleReader
```

```python
path = '/tmp/pixc/SWOT_L2_HR_PIXC_019_113_231L_20240803T194540_20240803T194551_PIC0_01_swot_66bcd402-2ba9-3cf9-a471-28d6d10297ed_V1-SWOT_L2_HR_PIXC_019_113_231L_20240803T194540_20240803T194551_PIC0_01_swot_66bcd402-2ba9-3cf9-a471-28d6d10297ed_V1/SWOT_L2_HR_PIXC_019_113_231L_20240803T194540_20240803T194551_PIC0_01.nc'

ncsimple = NcSimpleReader(path)#, ['height', 'sig0', 'classification'])

```

```python
ncsimple.open_dataset()
ncsimple.data
```

```python
# show some information
str_time_start, dt_time_start, cycle_number, pass_number, tile_number, swath_size = (
    ncsimple.extract_info_from_nc_attrs(path)
)
print(dt_time_start, cycle_number, pass_number, tile_number)
```
